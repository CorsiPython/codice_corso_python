```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
SAMPLES = 201
START_X = 0
END_X = 21

X = np.linspace(START_X, END_X, SAMPLES)
Y = 2.5 * np.sin(X) + X + np.random.randn(SAMPLES) * 2.0
```

```python
plt.scatter(X, Y)
```

$ y(x) = \sin(x) + x + \epsilon $

$\hat y (x) = \mu_x + \theta x$

```python
import pandas as pd


df = pd.DataFrame(np.vstack([X, Y]).T, columns=["x", "y"])
```

```python
X_pred = np.linspace(START_X - 10, END_X + 10, SAMPLES + 100)
y_hat = 0 + X_pred * 1.0
```

```python
X.shape
```

```python
y_hat.shape
```

```python
plt.scatter(X, Y)
plt.plot(X_pred, y_hat, color="orange")

plt.grid()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model
```

```python
def preprocess(X):
    return np.vstack([X, np.ones_like(X), np.sin(X)]).T
```

```python
preprocess(X).shape
```

```python
model.fit(preprocess(X), Y)
```

```python
X_pred.shape
```

```python
prezzo = model.predict(preprocess(X_pred))
```

```python
plt.scatter(X, Y)
plt.plot(X_pred, prezzo, color="orange")

plt.grid()
```
