## Python vs Numpy

```python
import time

arr = np.arange(30_000_000)
start = time.time()
result = [x**2 for x in arr]
print(f"Loop: {time.time() - start:.2f}s")
start = time.time()
result = arr**2
print(f"NumPy: {time.time() - start:.2f}s")
```

## Esercizio 1: Temperature

```python
temp = np.array([22, 25, 28, 26, 23, 27, 29])

media = temp.mean()
massima = temp.max()
minima = temp.min()

print(media, massima, minima)
print(np.sum((temp > 25)))
print(np.argmax(temp) + 1)
```
