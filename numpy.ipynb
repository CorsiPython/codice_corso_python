```python
X = np.array([[1.1, 2.1, 2.9, 6.0], [2, 3, 4, 1]]).T
X = np.vstack([X[:, 0], X[:, 1], np.cos(X[:, 0])]).T
y = np.array([2, 4, 6, 3]) + 2.0 * np.cos(X[:, 0])
```

```python
X
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

# 3D scatter of (x1, x2, y)
fig = plt.figure(figsize=(6, 5))
ax = fig.add_subplot(111, projection="3d")
x1 = X[:, 0]
x2 = X[:, 1]
ax.scatter(x1, x2, y, c="r", marker="o")
ax.set_xlabel("x1")
ax.set_ylabel("x2")
ax.set_zlabel("y")
ax.set_title("3D Scatter of data (x1, x2, y)")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Build design matrix with intercept and compute pseudoinverse solution
A = np.column_stack([np.ones(X.shape[0]), X])
w = np.linalg.pinv(A) @ y
print("Coefficients (intercept, b1, b2):", w)
# Create grid and evaluate plane
x1_grid, x2_grid = np.meshgrid(
    np.linspace(X[:, 0].min() - 1, X[:, 0].max() + 1, 20),
    np.linspace(X[:, 1].min() - 1, X[:, 1].max() + 1, 20),
)
y_grid = w[0] + w[1] * x1_grid + w[2] * x2_grid + w[3] * x1_grid
fig = plt.figure(figsize=(7, 6))
ax = fig.add_subplot(111, projection="3d")
# plot original points
ax.scatter(X[:, 0], X[:, 1], y, color="k")
# plot fitted plane
ax.plot_surface(x1_grid, x2_grid, y_grid, alpha=0.6, cmap="viridis")
ax.set_xlabel("x1")
ax.set_ylabel("x2")
ax.set_zlabel("y")
ax.set_title("Least-squares plane (pseudoinverse)")
plt.show()
```
