```python
x = np.linspace(0, 2 * np.pi, 6)
y = np.sin(x)
fig, ax = plt.subplots()
y2 = np.cos(x)
ax.scatter(x, y, color="#41bb4d", linewidth=2, label="sin(x)")
ax.scatter(x, y2, color="orange", linewidth=2, label="cos(x)")
ax.legend()
ax.grid()
```

```python
fig, ax = plt.subplots()
ax.scatter(
    df["statura"],
    df["eta"],
    c=df["sesso"].map({"F": "#e15759", "M": "#4e79a7"}),
    alpha=0.7,
)
ax.set_xlabel("Altezza (cm)")
ax.set_ylabel("EtÃ  (anni)")
plt.grid()
plt.legend()
plt.show()
```

```python
categorie = ["A", "B", "C"]
valori = [120, 90, 150]
fig, ax = plt.subplots()
plt.grid(axis="y", alpha=0.8, zorder=0)
ax.bar(categorie, valori, color="#59a14f", zorder=2)
ax.set_ylabel("Fatturato (kâ¬)")
```
